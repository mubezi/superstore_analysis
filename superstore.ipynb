```python
topcity_sales=data.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10)
topcity_sales
```

```python
# best cities by sales
```

```python
city_sales=data.groupby("City")['Sales'].sum().sort_values(ascending=False).head(10).reset_index()
city_sales
```

```python
sns.barplot(data=city_sales,x='City',y = 'Sales')
plt.xticks(rotation='vertical')
plt.title('Top cities by sales')
plt.show()
```

```python
#worst cities by sales
```

```python
worst_cities_sales=data.groupby("City")["Sales"].sum().sort_values().head(10).reset_index()
worst_cities_sales
```

```python
sns.barplot(data=worst_cities_sales,x='City',y='Sales',errorbar=None)
plt.xticks(rotation='vertical')
plt.title('least performing cities by sales')
plt.show()
```

```python
#least and most profitable cities
least_profit=data.groupby('City')["Profit"].sum().sort_values().head(10).reset_index()
least_profit
```

```python
most_profit=data.groupby('City')["Profit"].sum().sort_values(ascending=False).head(10).reset_index()
most_profit
```

```python
fig, axes = plt.subplots(1,2,figsize=(12,5))
sns.barplot(ax=axes[0],data=most_profit,x='City',y='Profit')
axes[0].tick_params(axis='x', rotation=90)
axes[0].set_title('Most profitable cities')

sns.barplot(ax=axes[1],data=least_profit,x='City',y='Profit')
axes[1].tick_params(axis='x', rotation =90)
axes[1].set_title('Least profitable cities')
plt.show()
```

```python
#most and least saleable state
```

```python
data['State'].unique()
```

```python
most_state_sales=data.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10).reset_index()
most_state_sales
```

```python
least_state_sales=data.groupby('State')['Sales'].sum().sort_values(ascending=True).head(10).reset_index()
least_state_sales
```

```python
fig,axes = plt.subplots(1,2,figsize=(12,5))

sns.barplot(ax=axes[0],data = most_state_sales, x='State',y = 'Sales')
axes[0].tick_params(axis ='x' ,rotation=90)
axes[0].set_title('Most saleable states')
sns.barplot(ax=axes[1],data = least_state_sales, x='State',y = 'Sales')
axes[1].tick_params(axis ='x' ,rotation=90)
axes[1].set_title('Least saleable states')
plt.show()
```

```python
#least and most profitable states
```

```python
most_state_profit=data.groupby('State')['Profit'].sum().sort_values(ascending=False).reset_index().head(10)
most_state_profit
```

```python
least_state_profit=data.groupby('State')['Profit'].sum().sort_values(ascending=True).head(10).reset_index() #either way only these ten states have negative profits
least_state_profit
```

```python
fig,axes = plt.subplots(1,2, figsize=(12,5))

sns.barplot(ax=axes[0], data = most_state_profit, x='State', y='Profit')
axes[0].tick_params(axis='x', rotation=90)
axes[0].set_title('Most profitable states')

sns.barplot(ax=axes[1], data = least_state_profit, x='State', y='Profit')
axes[1].tick_params(axis='x', rotation=90)
axes[0].set_title('Least profitable states')

plt.show()
```

```python
#profit verus sales for state
```

```python
state_profit_sales=data.groupby('State')['Profit','Sales'].sum().sort_values(by='Sales',ascending=False).reset_index().head(20)
state_profit_sales
```

```python
state_profit_sales.plot(x="State",y=["Sales","Profit"], kind='bar')
plt.title("Sales vs Profit per State.(Top 20 saleable cities)")
plt.show()
```

```python
#regions
```

```python
region_sum=data.groupby('Region').sum().reset_index().round(2)
region_sum
```

```python
fig, axes = plt.subplots(1,3, figsize=(19,5))

sns.barplot(ax=axes[0], data=region_sum, x='Region', y='Sales',errorbar=None)
axes[0].set_title('Sales vs Region')
sns.barplot(ax=axes[1], data=region_sum, x='Region', y='Profit',errorbar=None)
axes[1].set_title('Profit vs Region')
sns.barplot(ax=axes[2], data=region_sum, x='Region', y='Quantity',errorbar=None)
axes[2].set_title('Quantity vs Region')
plt.show()
```
